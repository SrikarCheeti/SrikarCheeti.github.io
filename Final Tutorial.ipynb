# <center>Introduction</center>
#### Rocket League is an online multiplayer game, which can be described with 2 words, "Car Soccer". There are multiple modes, with varying differences such as different balls, or team sizes, however for the purpose of this project we will focus on the standard 3v3 mode, which is also the mode professional Rocket League is played. 
#### The game is played like soccer, where the ball starts in the middle, however each team will get the ability to move at the same time and will be placed equally far from the ball. The two major mechanics to Rocket League is boosting and jumping, a mechanic is something the player can do in the game to gain an advantage. 
#### If the player's car has all 4 wheels touching the ground, the player is able to jump, then for a short time while in the air, they are able to flip in any direction they want. This mechanic provides a speed increase, by repeated flipping in the desired direction, over just driving. 
#### The second mechanic is boosting, which can be understood very easily with the words,"Rocket Booster". This is essentially what boosting is, the player has a capacity of 100 boost, and by clicking a button they activate a booster on the back of their car that pushes them forwards. As they use boost, they continuously loose it from their capacity, they can regain boost by driving over boost pads placed around the field. 
#### Just like soccer, the team that scores the most wins, this can be done through good defense or dominant offense. If the game is tied at the end of the game, the game goes to overtime, where the next team to score wins. There are some more complicated mechanics, and play styles, however this is the base of the game. RLCS stands for Rocket League Championship Series, throughout this project RLCS refers to professional rocket league.

#### My next thought was to find the unique values and make a dictonary of team_id to team_name.This way I do not need to access the website link everytime a team is repeated. df.apply() works significantly faster than loops, If i can get a function or dict that is able to do what I want, it is much faster than using a loop.


```python
from bs4 import BeautifulSoup
import numpy as np

#Dict of team_id to team name
team_id_dict = {}
team_ids = major_table["team_id"].unique()
#Get each team name
for i in range(len(team_ids)):
    #Get page we want
    curr_id = np.take(team_ids,i)
    str_id = str(curr_id)
    page_name = "".join(["https://zsr.octane.gg/teams/", str_id])
    page = requests.get(page_name)
    #Get html of it
    soup = BeautifulSoup(page.content, 'html.parser')
    soup = soup.prettify()
    #Find substring we want
    index_start = soup.find("name") + 7
    index_end = soup.find("\"",index_start)
    #Add to dict
    team_id_dict[str_id] = soup[index_start:index_end]
    
#Replace old team_id as new team names
major_table["team_id"] = major_table["team_id"].apply(lambda cell: team_id_dict[cell])

display(major_table)
```

#### Some data that will be of interest later on are dates. Over this entire dataset, the timeframe is from 2021-2022 because the dataset is RLCS 21-22. Adding the date to each row could help later on with graphing player performance based on chosen metrics, and predicting future performance.

#### !!!!CAUTION THIS BLOCK TAKES AROUND 10 MINS TO RUN!!!!
#### With ~5000 unique game_ids, this loop must go over all 5000 games to find the dates

#### I tried to reduce the amount of loops the for loop goes through, however if there are 5000 game_ids, I can't know the date unless I check which takes time. This method is an improvement from iterating over the entire dataframe, however it is still slower. This is slower than the previous method with teams because there are only a few 100 teams, however 5000 game ids.

```python
i = 0
game_id_dict = {}
game_ids = major_table["game_id"].unique()
#Get each date for each unique game
for curr_id in game_ids:
    #Get page address
    str_id = str(curr_id)
    page_name = "".join(["https://zsr.octane.gg/games/", str_id])
    page = requests.get(page_name)
    #Get the html from the page
    soup = BeautifulSoup(page.content, 'html.parser')
    soup = soup.prettify()
    #Find the substring we want
    index_start = soup.find("date") + 7
    index_end = soup.find("\"",index_start)
    #Add substring to dict
    game_id_dict[str_id] = soup[index_start:index_end]
    
#.apply() works significantly faster than loops 
major_table["date"] = major_table["game_id"].apply(lambda cell: game_id_dict[cell])
```

```python
#Now that we have a date column, lets convert it to a datetime type instead so we can use it as such later on
from datetime import datetime
#Turn into a string to manipulate the date values
major_table = major_table.astype({"date":'string'})
#Remove columns with date as "stamp"
major_table = major_table.drop(major_table[(major_table['date'] == "stamp")].index)
#Reshape %Y-%m-%dT%H:%M:%SZ to delete unnecessary T and Z
#.apply() works significantly faster than loops and is significantly easier to use/write
major_table["date"] = major_table["date"].apply(lambda cell: cell.replace('T', ' ', 1))
major_table["date"] = major_table["date"].apply(lambda cell: cell.replace('Z', '', 1))
#Convert to datetime object
major_table["date"] = major_table["date"].apply(lambda cell: datetime.strptime(cell, '%Y-%m-%d %H:%M:%S'))
```

#### For the next sections I plot the 2 metrics I want, then make a line of best fit. I do this to see how the data is laid out, then the regression line is a visual representation of the average.
```python
import matplotlib.pyplot as plt
#Making a scatter plot with a line of best fit
plt.scatter(major_table["core_goals"],major_table["core_score"])
m, b = np.polyfit(major_table["core_goals"],major_table["core_score"], 1)
plt.plot(major_table["core_goals"], m*major_table["core_goals"]+b)
plt.xlabel('Goals Scored')
plt.ylabel('Score Gained')
print("slope: " + str(m))
print("intercept: " + str(b))
```

# Hypothesis Testing

#### Advanced Rating is an overall rating of a player's performance. It is convienent because it considers all factors and variables to determine the rating. If we can use multiple Spearman's Test to find the strength and direction of a relationship between some chosen variables, and advanced_rating.

#### Then with this information we can find out the variables with the highest strength and direction to advanced_rating and consider those variables as the "most important". With an idea of which variables are more important, we can then use this information to confirm or deny thoughts on the roster moves SSG made. Such as based on Sypical's performance should they have tried to keep him, or which free agent would have been the best pick up, and was Lj among the top of those free agents?

#### We are using Spearman's test because it fits with our data. "Spearman's coefficient is appropriate for both continuous and discrete ordinal variables." https://archive.org/details/jmpforbasicuniva00leha. It also tells us p-value between 2 variables, which is useful to tell us if we can reject the null hypothesis. I am unsure of what to consider our data, continuous or discrete ordinal variables. Something like advanced rating has a scale where the rating can be any number for 0 to infinity. However something like goals is definitely a discrete ordinal variable, since there is an order, but only a certain number of goals. Someone cannot have 2.5 goals, Spearman's test is convinent because it can accommodate both varaible types which is why I chose it.


```python
from scipy import stats

#The Null hypothesis is there is no relation between goals and advanced_rating

stats.spearmanr(major_table["core_goals"],major_table["advanced_rating"])
```

#### Since the p value is very low(so low it was rounded to 0), we can reject the Null hypothesis that there is no relation between goals and advanced_rating. We also know there is a .7966 correlation
